```python
from sentence_transformers import SentenceTransformer

#모델 불러오기
model = SentenceTransformer('distiluse-base-multilingual-cased-v1') # 문장 및 단락을 512차원 밀집 벡터 공간에 매핑하고 클러스터링 또는 의미 검색과 같은 작업에 사용할 수 있습니다.
model_1 = SentenceTransformer('bert-base-nli-mean-tokens') # 기본 문장변환기
```

```python
import pandas as pd
```

```python
address = pd.read_csv('/content/drive/MyDrive/DTS_AI_TEST_DATA.csv', encoding='MS949')
#address
addr_data = address["ADDR"]

addr = addr_data.values
addr
```

```python
sentence_embeddings = model.encode(addr, convert_to_tensor=True)

sentence_embeddings_1 = model_1.encode(addr, convert_to_tensor=True)

print('Sample BERT embedding vector - length', len(sentence_embeddings[0]))
```

```python
#@title 기본 제목 텍스트
import scipy

query_addr = '\uB3C4\uBD09\uB85C 17' #@param {type: 'string'}

queries = [query_addr]
query_embeddings = model.encode(queries)

query_embeddings_1 = model_1.encode(queries)

number_top_matches = 5 #@param {type: "number"}

print("Semantic 검색 결과")

for query, query_embedding in zip(queries, query_embeddings):
    distances = scipy.spatial.distance.cdist([query_embedding], sentence_embeddings, "cosine")[0]

    results = zip(range(len(distances)), distances)
    results = sorted(results, key=lambda x: x[1])

    print("\n\n======================\n\n")
    print("검색어:", query)
    print("\n가장 유사한 문장 Top 5 :")

    for idx, distance in results[0:number_top_matches]:
        print(addr[idx].strip(), "(정확도: %.4f)" % (1-distance))

for query, query_embedding in zip(queries, query_embeddings_1):
    distances = scipy.spatial.distance.cdist([query_embedding], sentence_embeddings_1, "cosine")[0]

    results = zip(range(len(distances)), distances)
    results = sorted(results, key=lambda x: x[1])

    print("\n\n========22222222222222222==============\n\n")
    print("검색어:", query)
    print("\n가장 유사한 문장 Top 5 :")

    for idx, distance in results[0:number_top_matches]:
        print(addr[idx].strip(), "(정확도: %.4f)" % (1-distance))
```
